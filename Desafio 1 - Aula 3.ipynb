```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
print(cv2.__version__)

#Pop-up do Matplotlib
#%matplotlib qt

#Exibição na mesma tela do Jupyter
%matplotlib inline

plt.style.use('seaborn')
sns.set_style("whitegrid", {'axes.grid' : False})
```

```python
imagem = cv2.imread("robocop.jpg", cv2.IMREAD_GRAYSCALE)
borda_canny = cv2.Canny(imagem, 180, 180)
ret, imagem_limiarizada = cv2.threshold(borda_canny, 0, 255, cv2.THRESH_BINARY_INV)
plt.figure(figsize=(15,15))
plt.imshow(imagem_limiarizada, cmap="gray")
```
