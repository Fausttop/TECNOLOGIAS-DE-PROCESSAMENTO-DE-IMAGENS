```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
print(cv2.__version__)

#Pop-up do Matplotlib
#%matplotlib qt

#Exibição na mesma tela do Jupyter
%matplotlib inline

#plt.rcParams["figure.figsize"]\
#(8,5)

plt.style.use('seaborn')
sns.set_style("whitegrid", {'axes.grid' : False})
```

```python
baralho = cv2.imread("imagens/playing-cards.jpg")
baralho = cv2.cvtColor(baralho, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.imshow(baralho)
```

```python
copas7 = cv2.imread("imagens/playing-cards-7-copas.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(copas7, cmap='gray')
```

```python
baralho_gray = cv2.cvtColor(baralho, cv2.COLOR_BGR2GRAY)

correspondencia_template = cv2.matchTemplate(baralho_gray, copas7, cv2.TM_CCOEFF_NORMED)
w, h = copas7.shape[::-1]

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(correspondencia_template)

top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)

cv2.rectangle(baralho, top_left, bottom_right, (107,35,142), 3)

plt.figure(figsize=(20,20))
plt.imshow(baralho)
```
